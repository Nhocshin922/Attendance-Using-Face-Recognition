```python
from ArcFace.mobile_model import mobileFaceNet
import torch as t
from PIL import Image
import numpy as np
import cv2
from torchsummary import summary
```

```python
saved_model = "C:\\Users\\ACER\\Documents\\InternProjects\\Attendant Recognition\\Face_Recognition_using_pytorch-master\\Face_Recognition_using_pytorch-master\\ArcFace\\model\\069.pth"
threshold = 0.30896
```

```python
# Check if CUDA is available
device = t.device('cuda' if t.cuda.is_available() else 'cpu')

# Load the model
model = mobileFaceNet()
checkpoint = t.load(saved_model, map_location=device)  
if 'backbone_net_list' in checkpoint:
    model.load_state_dict(checkpoint['backbone_net_list'])
else:
    model.load_state_dict(checkpoint)

model.eval().to(device)

summary(model, input_size=(3, 112, 112))
```

```python
# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('failed to open camera!!!')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print('failed to grab frame')
        break

    # Convert BGR to grayscale (Haar Cascade works better with grayscale images)
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces using Haar Cascade
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Draw bounding boxes
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 155, 255), 2)

    cv2.imshow('img', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
